```python
# Fetch and display the status of the created pull request
if response.status_code == 201:
    pr_number = response.json()["number"]
    pr_status_url = f"https://api.github.com/repos/wickedyoda/Glinet_discord_bot/pulls/{pr_number}"
    pr_status_resp = requests.get(pr_status_url, headers=headers)
    if pr_status_resp.status_code == 200:
        pr_info = pr_status_resp.json()
        print(f"PR #{pr_info['number']} status: {pr_info['state']}")
        print(f"Title: {pr_info['title']}")
        print(f"URL: {pr_info['html_url']}")
    else:
        print(f"Failed to fetch PR status: {pr_status_resp.status_code}")
else:
    print("No PR to check status for.")
```
